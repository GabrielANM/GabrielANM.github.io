```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plot_objetc(objetc, new=True, color='b'):
  x = [p[0] for p in objetc]
  y = [p[1] for p in objetc]

  if new:
    plt.figure(figsize=(15,15))
  plt.plot(x, y, c=color);
```
